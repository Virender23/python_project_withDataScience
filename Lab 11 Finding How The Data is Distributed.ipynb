### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
import pandas as pd


# Display the column names and data types
print("Column Names and Data Types:")
print(df.dtypes)

# Display a summary of the dataset
print("\nSummary Information:")
print(df.info())

# Display the first few rows to get an initial sense of the data
print("\nFirst 5 Rows of the Dataset:")
print(df.head())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# Identify missing values
missing_data = df.isnull().sum()

# Display missing values per column
print("Missing Values per Column:")
print(missing_data)

# Percentage of missing data in each column
missing_percentage = (df.isnull().mean() * 100)
print("\nPercentage of Missing Data per Column:")
print(missing_percentage)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here

# Analyze key columns: Employment, JobSat, and YearsCodePro
print("Value Counts for Employment:")
print(df['Employment'].value_counts())

print("\nValue Counts for JobSat (Job Satisfaction):")
print(df['JobSat'].value_counts())

print("\nValue Counts for YearsCodePro (Professional Coding Experience):")
print(df['YearsCodePro'].value_counts())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
# Plot a pie chart for JobSat (Job Satisfaction)
job_sat_counts = df['JobSat'].value_counts()

plt.figure(figsize=(7, 7))
job_sat_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff9999'])
plt.title('Job Satisfaction Distribution')
plt.ylabel('')  # Remove the label 'JobSat'
plt.show()
```

```python
import seaborn as sns

# Create a KDE plot (only useful if JobSat is numerical or ordinal)
plt.figure(figsize=(8, 6))
sns.kdeplot(df['JobSat'], shade=True, color='skyblue')
plt.title('Kernel Density Estimate for Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Density')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
!pip install matplotlib-venn
from matplotlib_venn import venn2


# Extract languages from the respective columns (assuming they are stored as comma-separated values)
languages_worked_with = set(df['LanguageHaveWorkedWith'].dropna().str.split(',').explode().str.strip())
languages_want_to_work_with = set(df['LanguageWantToWorkWith'].dropna().str.split(',').explode().str.strip())

# Create a Venn diagram
plt.figure(figsize=(8, 8))
venn2([languages_worked_with, languages_want_to_work_with], set_labels=('Have Worked With', 'Want to Work With'))
plt.title('Programming Languages Comparison: Have Worked With vs Want to Work With')
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

remote_by_region = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)

# Plotting the grouped bar chart
remote_by_region.plot(kind='bar', stacked=False, figsize=(10, 6), color=['#66b3ff', '#ff9999'])
plt.title('Remote Work Distribution by Country')
plt.ylabel('Number of Employees')
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work', labels=['No', 'Yes'])
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
print(df['YearsCodePro'].unique())

# Handle non-numeric values (e.g., 'Less than 1 year' -> 0)
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 51,  # Example: Replace with max value or based on your context
    '1-2 years': 1.5,  # Example: Use an average or an appropriate mapping
    # Add any other necessary mappings here...
})

# Convert 'YearsCodePro' to numeric values (if not already)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# If 'JobSat' is categorical, map it to numerical values (as before)
job_sat_mapping = {'Very Dissatisfied': 1, 'Dissatisfied': 2, 'Neutral': 3, 'Satisfied': 4, 'Very Satisfied': 5}
df['JobSat'] = df['JobSat'].map(job_sat_mapping)

# Calculate the Pearson correlation between JobSat and YearsCodePro
pearson_corr = df[['JobSat', 'YearsCodePro']].corr(method='pearson')

print("Pearson Correlation Matrix:")
print(pearson_corr)
```

```python
# Inspect the unique values in 'YearsCodePro' column
print(df['YearsCodePro'].unique())

# Replace non-numeric values in 'YearsCodePro' with appropriate numeric values
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,      # Convert 'Less than 1 year' to 0
    '1-2 years': 1.5,           # Convert '1-2 years' to the average of 1 and 2, or another appropriate value
    '3-5 years': 4,             # Replace '3-5 years' with an average value (e.g., 4)
    '6-10 years': 8,            # Replace '6-10 years' with an average value (e.g., 8)
    '11-20 years': 15,          # Replace '11-20 years' with an average value (e.g., 15)
    '21-30 years': 25,          # Replace '21-30 years' with an average value (e.g., 25)
    '31-40 years': 35,          # Replace '31-40 years' with an average value (e.g., 35)
    '41-50 years': 45,          # Replace '41-50 years' with an average value (e.g., 45)
    'More than 50 years': 51    # Replace 'More than 50 years' with 51 (or another appropriate value)
})

# Convert 'YearsCodePro' to numeric values (forcing errors to NaN)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# If 'JobSat' is categorical, map it to numerical values (as previously done)
job_sat_mapping = {'Very Dissatisfied': 1, 'Dissatisfied': 2, 'Neutral': 3, 'Satisfied': 4, 'Very Satisfied': 5}
df['JobSat'] = df['JobSat'].map(job_sat_mapping)

# Drop rows with NaN values in 'JobSat' or 'YearsCodePro' (if any)
df = df.dropna(subset=['JobSat', 'YearsCodePro'])

# Calculate the Spearman correlation between JobSat and YearsCodePro
spearman_corr = df[['JobSat', 'YearsCodePro']].corr(method='spearman')

print("Spearman Correlation Matrix:")
print(spearman_corr)
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
# Create the cross-tabulation between Employment and Education Level


employment_vs_edlevel = pd.crosstab(df['Employment'], df['EdLevel'])

# Debugging: Display the cross-tabulation and data types
print("Cross-tabulation Data:")
print(employment_vs_edlevel)

print("\nData Types:")
print(employment_vs_edlevel.dtypes)

# Ensure all values in the DataFrame are numeric
employment_vs_edlevel = employment_vs_edlevel.apply(pd.to_numeric, errors='coerce')

# Debugging: Confirm numeric conversion
print("\nAfter Conversion to Numeric:")
print(employment_vs_edlevel.dtypes)

# Plot the stacked bar chart
employment_vs_edlevel.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set3')

# Add title and labels
plt.title('Employment Status vs. Education Level')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
