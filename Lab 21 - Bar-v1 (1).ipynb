### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Check for any missing or NaN values in 'ConvertedCompYearly' column
df = df.dropna(subset=['ConvertedCompYearly'])

# Set up the plot
plt.figure(figsize=(10, 6))

# Plotting the histogram with seaborn
sns.histplot(df['ConvertedCompYearly'], bins=30, kde=True, color='blue')

# Adding titles and labels
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)', fontsize=14)
plt.xlabel('Yearly Compensation', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show plot
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Map the 'Age' categories to numeric values
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years and older': 70
}

# Apply the mapping to the 'Age' column
df['Age_numeric'] = df['Age'].map(age_mapping)

# Set up the plot
plt.figure(figsize=(10, 6))

# Plotting the boxplot with seaborn
sns.boxplot(x='Age_numeric', y='ConvertedCompYearly', data=df, palette="Set2")

# Adding titles and labels
plt.title('Box Plot of Yearly Compensation by Age Group', fontsize=14)
plt.xlabel('Age Group (Numeric)', fontsize=12)
plt.ylabel('Yearly Compensation', fontsize=12)

# Show plot
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Set up the plot
plt.figure(figsize=(10, 6))

# Plotting the horizontal bar chart using seaborn
sns.countplot(y='MainBranch', data=df, palette='Set2', order=df['MainBranch'].value_counts().index)

# Adding titles and labels
plt.title('Distribution of Respondents\' Primary Roles (MainBranch)', fontsize=14)
plt.xlabel('Number of Respondents', fontsize=12)
plt.ylabel('Primary Role (MainBranch)', fontsize=12)

# Show plot
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Split the 'LanguageWantToWorkWith' by semicolon and count the occurrence of each language
languages = df['LanguageWantToWorkWith'].dropna().str.split(';', expand=True).stack()

# Count the frequencies of each programming language
language_counts = languages.value_counts().head(5)

# Set up the plot
plt.figure(figsize=(10, 6))

# Plotting the vertical bar chart using seaborn
sns.barplot(x=language_counts.index, y=language_counts.values, palette='Set2')

# Adding titles and labels
plt.title('Top 5 Programming Languages Respondents Want to Work With', fontsize=14)
plt.xlabel('Programming Languages', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)

# Show plot
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Map the 'Age' categories to numeric values if not already done
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years and older': 70
}

# Apply the mapping to the 'Age' column (if not already applied)
df['Age_numeric'] = df['Age'].map(age_mapping)

# Group by 'Age_numeric' and calculate the median for JobSatPoints_6 and JobSatPoints_7
median_job_sat = df.groupby('Age_numeric')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Set up the plot
plt.figure(figsize=(12, 8))

# Plotting the stacked bar chart
median_job_sat.plot(kind='bar', stacked=True, color=['skyblue', 'orange'], figsize=(12, 8))

# Adding titles and labels
plt.title('Median Job Satisfaction by Age Group (JobSatPoints_6 & JobSatPoints_7)', fontsize=14)
plt.xlabel('Age Group (Numeric)', fontsize=12)
plt.ylabel('Median Job Satisfaction Points', fontsize=12)

# Show plot
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Split the 'DatabaseHaveWorkedWith' by semicolon and count the occurrence of each database
databases = df['DatabaseHaveWorkedWith'].dropna().str.split(';', expand=True).stack()

# Count the frequencies of each database
database_counts = databases.value_counts()

# Set up the plot
plt.figure(figsize=(12, 8))

# Plotting the bar chart using seaborn
sns.barplot(x=database_counts.index, y=database_counts.values, palette='Set2')

# Adding titles and labels
plt.title('Most Commonly Used Databases Among Respondents', fontsize=14)
plt.xlabel('Database', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show plot
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Map the 'Age' categories to numeric values if not already done
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years and older': 70
}

# Apply the mapping to the 'Age' column (if not already applied)
df['Age_numeric'] = df['Age'].map(age_mapping)

# Group by 'Age_numeric' and calculate the median of 'ConvertedCompYearly'
median_compensation = df.groupby('Age_numeric')['ConvertedCompYearly'].median().reset_index()

# Set up the plot
plt.figure(figsize=(10, 6))

# Plotting the grouped bar chart using seaborn
sns.barplot(x='Age_numeric', y='ConvertedCompYearly', data=median_compensation, palette='Set2')

# Adding titles and labels
plt.title('Median Compensation Across Different Age Groups', fontsize=14)
plt.xlabel('Age Group (Numeric)', fontsize=12)
plt.ylabel('Median Compensation (ConvertedCompYearly)', fontsize=12)

# Show plot
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataset named df
# If the data is in CSV, you can load it like this:
# df = pd.read_csv('your_data.csv')

# Count the number of respondents by country
country_counts = df['Country'].value_counts()

# Set up the plot
plt.figure(figsize=(12, 8))

# Plotting the bar chart using seaborn
sns.barplot(x=country_counts.index, y=country_counts.values, palette='Set2')

# Adding titles and labels
plt.title('Distribution of Respondents by Country', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha='right')

# Show plot
plt.show()
```
