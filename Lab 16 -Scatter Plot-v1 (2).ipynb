#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
# Mapping of Age groups to numerical values
age_mapping = {
    'Under 18 years old': 17, 
    '18-24 years old': 21, 
    '25-34 years old': 30, 
    '35-44 years old': 40, 
    '45-54 years old': 50, 
    '55-64 years old': 60, 
    '65 years and older': 70
}

# Convert Age column into numerical values
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert JobSatPoints_6 to numeric, errors='coerce' will convert invalid values to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows where AgeNumeric or JobSatPoints_6 is NaN
df_cleaned = df.dropna(subset=['AgeNumeric', 'JobSatPoints_6'])

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['AgeNumeric'], df_cleaned['JobSatPoints_6'], alpha=0.7, color='b')

# Add labels and title
plt.title('Age vs. Job Satisfaction (JobSatPoints_6)', fontsize=14)
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Display the plot
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
# Convert JobSatPoints_6 to numeric, errors='coerce' will convert invalid values to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows where ConvertedCompYearly or JobSatPoints_6 is NaN
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['ConvertedCompYearly'], df_cleaned['JobSatPoints_6'], alpha=0.7, color='g')

# Add labels and title
plt.title('Yearly Compensation vs. Job Satisfaction (JobSatPoints_6)', fontsize=14)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Display the plot
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
import seaborn as sns
# Mapping of Age groups to numerical values
age_mapping = {
    'Under 18 years old': 17, 
    '18-24 years old': 21, 
    '25-34 years old': 30, 
    '35-44 years old': 40, 
    '45-54 years old': 50, 
    '55-64 years old': 60, 
    '65 years and older': 70
}

# Convert Age column into numerical values
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert JobSatPoints_6 to numeric, errors='coerce' will convert invalid values to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows where AgeNumeric or JobSatPoints_6 is NaN
df_cleaned = df.dropna(subset=['AgeNumeric', 'JobSatPoints_6'])

# Set seaborn style
sns.set(style="whitegrid")

# Create the scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='AgeNumeric', y='JobSatPoints_6', data=df_cleaned, scatter_kws={'alpha':0.7, 'color':'b'}, line_kws={'color':'red'})

# Add labels and title
plt.title('Age vs. Job Satisfaction with Regression Line', fontsize=14)
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Display the plot
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
# Mapping of Age groups to numerical values
age_mapping = {
    'Under 18 years old': 17, 
    '18-24 years old': 21, 
    '25-34 years old': 30, 
    '35-44 years old': 40, 
    '45-54 years old': 50, 
    '55-64 years old': 60, 
    '65 years and older': 70
}

# Convert Age column into numerical values
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Drop rows where AgeNumeric or YearsCodePro is NaN
df_cleaned = df.dropna(subset=['AgeNumeric', 'YearsCodePro'])

# Set seaborn style
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AgeNumeric', y='YearsCodePro', data=df_cleaned, alpha=0.7, color='g')

# Add labels and title
plt.title('Age vs. Work Experience (YearsCodePro)', fontsize=14)
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Work Experience (YearsCodePro)', fontsize=12)

# Display the plot
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
 #Mapping of Age groups to numerical values
age_mapping = {
    'Under 18 years old': 17, 
    '18-24 years old': 21, 
    '25-34 years old': 30, 
    '35-44 years old': 40, 
    '45-54 years old': 50, 
    '55-64 years old': 60, 
    '65 years and older': 70
}

# Convert Age column into numerical values
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Convert JobSatPoints_6 to numeric, errors='coerce' will convert invalid values to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows where ConvertedCompYearly, JobSatPoints_6, or AgeNumeric is NaN
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNumeric'])

# Use .loc to avoid SettingWithCopyWarning
df_cleaned.loc[:, 'BubbleSize'] = df_cleaned['AgeNumeric'] * 20  # Scaling factor for bubble size

# Set seaborn style for better visualization
sns.set(style="whitegrid")

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(x=df_cleaned['ConvertedCompYearly'], 
            y=df_cleaned['JobSatPoints_6'], 
            s=df_cleaned['BubbleSize'],  # Size of bubbles
            alpha=0.6, 
            color='b', edgecolors='w', linewidth=0.5)

# Add labels and title
plt.title('Compensation vs. Job Satisfaction with Age as Bubble Size', fontsize=14)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Display the plot
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here

# Grouping by language and calculating the average JobSatPoints_6 and count of each language
language_stats = df.groupby('LanguageHaveWorkedWith').agg(
    avg_job_satisfaction=('JobSatPoints_6', 'mean'),
    count=('JobSatPoints_6', 'size')
).reset_index()

# Set seaborn style for better visualization
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='count', y='avg_job_satisfaction', data=language_stats, s=100, color='b', edgecolors='w', linewidth=0.5)

# Add labels and title
plt.title('Job Satisfaction vs. Popularity of Programming Languages', fontsize=14)
plt.xlabel('Number of Respondents (Popularity)', fontsize=12)
plt.ylabel('Average Job Satisfaction', fontsize=12)

# Display the plot
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
# Set seaborn style for better visualization
sns.set(style="whitegrid")

# Create the scatter plot with color coding for Employment types
plt.figure(figsize=(10, 6))

sns.scatterplot(x='ConvertedCompYearly', 
                y='JobSatPoints_6', 
                hue='Employment',  # Color by Employment type
                data=df, 
                palette='Set1',  # Set a color palette
                s=100, 
                edgecolors='w', 
                linewidth=0.5)

# Add labels and title
plt.title('Compensation vs. Job Satisfaction by Employment Type', fontsize=14)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Show legend for employment types
plt.legend(title='Employment Type', loc='upper right')

# Display the plot
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
# Convert Age into numerical values for plotting
age_mapping = {
    'Under 18 years old': 17, 
    '18-24 years old': 21, 
    '25-34 years old': 30, 
    '35-44 years old': 40, 
    '45-54 years old': 50, 
    '55-64 years old': 60, 
    '65 years and older': 70
}

df['AgeNumeric'] = df['Age'].map(age_mapping)

# Set seaborn style for better visualization
sns.set(style="whitegrid")

# Create the scatter plot with color coding for countries and markers for age groups
plt.figure(figsize=(10, 6))

sns.scatterplot(x='YearsCodePro', 
                y='AgeNumeric', 
                hue='Country',  # Color by Country
                style='Age',  # Use different markers for each Age group
                data=df, 
                palette='Set1',  # Color palette for countries
                markers=['o', 's', 'D', '^', 'v', '<', '>', 'p'],  # Added one more marker ('p')
                s=100, 
                edgecolors='w', 
                linewidth=0.5)

# Add labels and title
plt.title('Work Experience vs. Age Group by Country', fontsize=14)
plt.xlabel('Years of Work Experience (YearsCodePro)', fontsize=12)
plt.ylabel('Age Group (Age)', fontsize=12)

# Show legend for countries and age groups
plt.legend(title='Country and Age Group', loc='upper left')

# Display the plot
plt.show()
```
